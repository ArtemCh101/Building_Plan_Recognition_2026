```python
import os
import shutil
import random
```

```python
def setup_yolo_dir(base_path):
    for folder in ['images', 'labels']:
        for split in ['train', 'val', 'test']:
            os.makedirs(os.path.join(base_path, folder, split), exist_ok=True)
```

```python
def split_dataset(src_folder, target_base, split_ratio=(0.8, 0.1, 0.1)):
    setup_yolo_dir(target_base)
    
    files = [f for f in os.listdir(src_folder) if f.endswith('.png')]
    random.seed(42)
    random.shuffle(files)
    
    train_idx = int(len(files) * split_ratio[0])
    val_idx = train_idx + int(len(files) * split_ratio[1])
    
    splits = {
        'train': files[:train_idx],
        'val': files[train_idx:val_idx],
        'test': files[val_idx:]
    }
    
    for split_name, file_list in splits.items():
        for f in file_list:
            label_f = f.replace('.png', '.txt')
            
            shutil.copy(os.path.join(src_folder, f), os.path.join(target_base, 'images', split_name, f))
            
            label_src = os.path.join(src_folder, label_f)
            if os.path.exists(label_src):
                shutil.copy(label_src, os.path.join(target_base, 'labels', split_name, label_f))
)
```

```python
split_dataset('data/colorful', 'yolo_data_colorful')
split_dataset('data/black_and_white', 'yolo_data_bw'
```
